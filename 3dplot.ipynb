```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('/home/singh/benchmarking/wandb_export_2024-10-03T00_43_36.248+05_30.csv')
```

```python
df = df[(df["epoch"] == 99) & (df["train_loss"] < 0.5) & (df["lr2"]<3) & (df["lr2"]>2)]
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf((-df['lr1']), (-df['lr2']), df['train_loss'], cmap="viridis", edgecolor="none")
ax.view_init(elev=90, azim=0, roll=0)

ax.set_xlabel('lr1')
ax.set_ylabel('lr2')
ax.set_zlabel('train_loss')
# ax.set_ylim(-2, -3)

plt.show()
```
